### Logical Operators

Nesne Yönelimli Programlama dillerinin hepsinde eşitlik, karşıtlık, büyüklük - küçüklük durumlarını kıyaslayabilmek için kullanılan tipik mantıksal eşitlemeler vardır. 
* '=' : Bu işaret matematikte eşittir gibi değil de. Atama operasyonlarında bir objeyi bir duruma bağlamak amacıyla kullanılır.
* '<' : Matematikteki küçüktür işareti.
* '>' : Matematikteki büyüktür işareti.
* '!=' : Eşit değildir işareti.
* '==' : Eşittir işareti

```python
a = 5
b = 10
if(a<b):
  print("A, B den küçüktür")

if(a!=b):
  print("A,B ile aynı değildir")

if(a==5):
  print("A = 5 tir.")
```

##Exception Handling

Biraz önce yaptığımız örneklerde de gördüğünüz gibi bazen gözümüzden bir şeyler kaçabiliyor kod yazarken.

Veya kullanıcıya değer verdirdiğimiz durumlarda bazen kullanıcı istediğimiz veri tipinde değer giremeyebiliyor.

Bunları engellemek için 'TRY'-'EXCEPT' blokları vardır.

Her 'Except' bloğu bulduğu en yakın 'Try' bloğunun içerisindeki duruma göre davranır.

Normalde çok fazla hata tipi ve bunlara karşı kullanılabilecek 'try' - 'exception' türleri vardır. Biz sadece en çok karşılaşabileceğiniz birkaç tanesini göstereceğiz.

Detaylı bilgi için : https://docs.python.org/3/library/exceptions.html ziyaret edebilirsin.


### ZeroDivisionError

```python
## Bu sadece bölme işlemlerinde değil de günlük hayatta döndürülen
## ve değeri sıfır olmaması gereken değerlerle karşılaşıldığında
## kullanılan exception tipidir. Genelde bu exceptiondan sonra
## gidip veri tekrardan kontrol edilir.
a =input()
b = input()
try: 
  print(int(a)/int(b))
except ZeroDivisionError:
  print('0 ile bölersen sonsuz olur')
```

### ValueError
